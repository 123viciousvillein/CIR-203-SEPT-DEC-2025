```python
# ===============================
# 1. Create a dictionary of 5 products and their stock quantities
# ===============================

inventory = {
    "Laptop": 15,
    "Mouse": 8,
    "Keyboard": 12,
    "Headphones": 5,
    "USB Cable": 20
}

print("1. Initial Inventory:")
print(inventory)

# ===============================
# 2. Add a new product and update the quantity of an existing one
# ===============================

inventory["Webcam"] = 18       # Adding new product
inventory["Mouse"] = 14        # Updating existing product quantity

print("\n2. After Adding and Updating:")
print(inventory)

# ===============================
# 3. Function to return products with stock less than 10
# ===============================

def low_stock_items(inv_dict):
    return [product for product, qty in inv_dict.items() if qty < 10]

low_stock = low_stock_items(inventory)

print("\n3. Products with Stock Less Than 10:")
print(low_stock)

# ===============================
# 4. Delete a discontinued product
# ===============================

del inventory["Headphones"]   # Removing discontinued product

print("\n4. After Deleting Discontinued Product:")
print(inventory)

# ===============================
# 5. Loop through using .items() and print product + quantity
# ===============================

print("\n5. Product List with Quantities:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
```
