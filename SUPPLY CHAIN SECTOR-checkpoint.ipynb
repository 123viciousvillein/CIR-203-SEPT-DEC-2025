```python
# ============================
# 1. Create a list of 10 delivery routes
# ============================

routes = [
    "Nairobi-Westlands",
    "Mombasa-Nyali",
    "Kisumu-CBD",
    "Nakuru-Lanet",
    "Eldoret-Town",
    "Nanyuki-Central",
    "Thika-Gatundu",
    "Naivasha-Industrial",
    "Meru-Makutano",
    "Nyeri-Kingongo"
]

print("1. Initial Routes (10 routes):")
print(routes)

# ============================
# 2. Append a new route and remove a discontinued one
# ============================

routes.append("Kitale-Central")  # New route added
routes.remove("Kisumu-CBD")     # Example discontinued route

print("\n2. After Adding and Removing Routes:")
print(routes)

# ============================
# 3. Sort the list alphabetically and reverse it
# ============================

routes.sort()       # Alphabetical
routes.reverse()    # Reverse order

print("\n3. Sorted and Reversed Routes:")
print(routes)

# ============================
# 4. Count how many routes start with the letter “N”
# ============================

count_N = sum(route.startswith("N") for route in routes)

print("\n4. Number of Routes Starting with 'N':", count_N)

# ============================
# 5. List comprehension: routes longer than 10 characters
# ============================

long_routes = [route for route in routes if len(route) > 10]

print("\n5. Routes Longer Than 10 Characters:")
print(long_routes)
```
