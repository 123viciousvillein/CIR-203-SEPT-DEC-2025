```python
# ============================
# 1. Create a tuple for a single patient
# ============================

patient = ("John Doe", 45, "120/80", 72)  # (Name, Age, Blood Pressure, Heart Rate)

print("1. Patient Tuple:")
print(patient)

# ============================
# 2. Access and print the patient’s age and heart rate
# ============================

age = patient[1]
heart_rate = patient[3]

print("\n2. Patient Age:", age)
print("   Patient Heart Rate:", heart_rate)

# ============================
# 3. Explanation: Why tuples are suitable?
# ============================

print("\n3. Explanation:")
print("Tuples are immutable, meaning their values cannot be changed accidentally.")
print("This makes them ideal for storing patient vitals that should remain as")
print("historical, unaltered medical records.")

# ============================
# 4. Convert tuple → list, update heart rate → tuple
# ============================

patient_list = list(patient)  # convert to list
patient_list[3] = 80          # update heart rate
updated_patient = tuple(patient_list)  # convert back to tuple

print("\n4. Updated Patient Tuple (heart rate changed):")
print(updated_patient)

# ============================
# 5. Create a tuple of 5 patients and extract all names
# ============================

patients = (
    ("John Doe", 45, "120/80", 72),
    ("Mary Ann", 50, "130/85", 78),
    ("Peter Kim", 39, "118/76", 70),
    ("Lucy Wanjiku", 60, "140/90", 82),
    ("Brian Otieno", 30, "110/70", 68)
)

# Extract names using list comprehension
names = [patient[0] for patient in patients]

print("\n5. Names of All Patients:")
print(names)
```
